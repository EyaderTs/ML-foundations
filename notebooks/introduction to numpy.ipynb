## Dot Product

```python

np.random.seed(0)
mat_1= np.random.randint(10,size = (5,3))
mat_2= np.random.randint(10,size = (5,3))
```

```python
mat_1.shape,mat_2.shape
```

```python
mat_1
```

```python
mat_2
```

```python
## Elemenet Wise Multiplication AKA - Hadamard product - which direct element by element multiplication
mat_1 * mat_2
```

```python
# now let us try the dot product
np.dot(mat_1,mat_2)
```

```python
mat_1
```

```python
# the above error is because of the shape dismatch of the arrays which affects the rule of dot products,
# so let us fix them out
# here is where the transpos comes to reshape the array so let us transpose shape of mat_2
mat_2.T
```

```python
mat_3 =  np.dot (mat_1,mat_2.T)
mat_3
```

```python
# because of (5,3) (3,5) = (5,5) as is shows it is taking the outer as shape, to do dot product the internal must be same.
mat_3.shape
```

## Dot product example (nut butter sales)

```python
np.random.seed(0)
#num of jars sold
sales_amount = np.random.randint(20, size = (5,3))
sales_amount
```

```python
weekly_sales = pd.DataFrame(sales_amount,index=["Mon","Tues","Wed","Thurs","Fri"], columns=["Almonde butter","Peanut butter", "Cashew butter"] )
weekly_sales
```

```python
weekly_sales.shape
```

```python
# create prices array
prices = np.array([10,8,12])
prices
```

```python
# create data frame for it
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Prices"],columns=["Almonde butter","Peanut butter", "Cashew butter"] )
```

```python
butter_prices
```

```python
butter_prices.shape
```

```python
weekly_sales.T.shape
```

```python
weekly_sales.T
```

```python
butter_prices
```

```python
butter_prices.shape
```

```python
weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
daily_sales.shape
```

```python
weekly_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T # transpose to much the shape
```

```python
weekly_sales
```
