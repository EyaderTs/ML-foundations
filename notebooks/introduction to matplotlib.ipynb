# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(5, 5))  # single plot, not a grid

#3. plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5. save and show 
fig.savefig("../images/sample-plot.png")
# plt.show()
```

# Subplots

### Two options to create subplots

```python
# subplot option 1 - using tuples
fig, ((ax1,ax2),(ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,5) )
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(),nut_butter_price.values());
ax4.hist(x);
plt.show()
```

```python
#subplot option 2 - indexing

fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(10,5) )
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values());
ax[1,1].hist(x);
plt.show()
```

```python
#let us try with other dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins = 30)
plt.show()
```

```python
# this plot is not ideal because each subplot scale is same
#this one use pyplot  method, plotting directly from pandas dataframe rather than oo method
heart_disease.plot.hist(figsize = (10,30),subplots= True)
plt.show()
```

```python
# which one i should use (pyplot or matplotlib oo method)?
# for quick plot okay to ouse the pyplot 
#for more advanced use OO method

heart_disease.head()
```

```python
over_50 = heart_disease[ heart_disease.age >50]
over_50.head()
```

```python
#pyplot method
over_50.plot(figsize=(10,6), kind= "scatter", x= "age", y = "chol", c= "target")
plt.show()
```

```python
#OO method mixed with pyplot method
fig,ax = plt.subplots(figsize = (10 ,6))
over_50.plot(kind= "scatter", x= "age", y = "chol", c= "target",ax=ax)
plt.show()
```

```python
#OO method mixed from scratch
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"], y = over_50["chol"], c=over_50["target"])

#customize plot
ax.set(title="Heart Disease and cholesterol levels", xlabel="age",ylabel="chol")

#add a legend
ax.legend (*scatter.legend_elements(),title= "Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = '--' )
plt.show()
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize= (10,10),sharex=True)

scatter0 = ax0.scatter(x=over_50["age"],y=over_50["chol"],c = over_50["target"])
ax0.set(title="Heart Disease and Cholesterol Levels",ylabel="Cholesterol")
ax0.axhline(y= over_50["chol"].mean(), linestyle= "--")
ax0.legend(*scatter0.legend_elements(),title='Target')

scatter1 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c = over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate", xlabel= "Age",ylabel="Cholesterol")
ax1.axhline(y= over_50["thalach"].mean(), linestyle= "--")
ax1.legend(*scatter1.legend_elements(),title='Target')

# add title for the whole figure
fig.suptitle("Heart Disease Analysis")

plt.show()
```
